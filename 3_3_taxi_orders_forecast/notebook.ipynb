## Задание 

### Описание проекта

Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час.

## Обучение

```python
grid_report = pd.DataFrame(columns=['model', 'rmse', 'params'])
```

```python
def my_train_test_split(data):
    train, test = train_test_split(data, shuffle=False, test_size=0.1,random_state=12345)
    return train, test
```

```python
def my_GridSearchCV(mode, parametrs, train_x, train_y):
    
    grid = GridSearchCV(model(), parametrs, cv=TimeSeriesSplit(), scoring='neg_root_mean_squared_error')
    grid.fit(train_x, train_y)
    
    row = [model.__name__, grid.best_score_  * -1, grid.best_params_]
    return pd.DataFrame([row], columns=['model', 'rmse', 'params'])
```

```python
train, test = my_train_test_split(data)
```

```python
max_lag = 30
```

```python
rolling_mean_size = 8
```

```python
def make_features(data):
    
    global max_lag
    global rolling_mean_size
    
    data.loc[:,'year'] = data.index.year
    data.loc[:,'month'] = data.index.month
    data.loc[:,'day'] = data.index.day
    data.loc[:,'dayofweek'] = data.index.dayofweek
    def my_GridSearchCV(mode, parametrs, train_x, train_y):
    
    grid = GridSearchCV(model(), parametrs, cv=TimeSeriesSplit(), scoring='neg_root_mean_squared_error')
    grid.fit(train_x, train_y)
    
    row = [model.__name__, grid.best_score_  * -1, grid.best_params_]
    return pd.DataFrame([row], columns=['model', 'rmse', 'params'])
    for lag in range(1, max_lag + 1):
        data.loc[:,'lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data.loc[:,'rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    return data.dropna()
```

```python
def split_x_y(data):
    x = data.drop('num_orders', axis=1)
    y = data['num_orders']
    return x, y
```

```python
def data_prepare(data):
    data = make_features(data)
    return split_x_y(data)
```

```python
train_x, train_y = data_prepare(train)
```

### Градиентный бустинг

```python
parametrs = {
    'num_leaves' :[31,21, 50, 70],
    'learning_rate': [0.4, 0.1],
    'n_estimators' : [50,100, 200, 500],
    'random_state':[12345]
}
```

```python
model = LGBMRegressor
result = my_GridSearchCV(model, parametrs,train_x,train_y)
```

```python
grid_report = grid_report.append(result)
```

```python
grid_report
```

### Простой регрессор

```python
model = DummyRegressor
```

```python
result = my_GridSearchCV(model, parametrs,train_x,train_y)
```

```python
grid_report = grid_report.append(result)
```

```python
grid_report
```

```python
parametrs = {'fit_intercept': ('True', 'False'), 'normalize' :('False','True')}
```

```python
model = LinearRegression
```

```python
result = my_GridSearchCV(model, parametrs,train_x,train_y)
```

```python
grid_report = grid_report.append(result)
```

```python
grid_report
```

### Решающие дерево

```python
parametrs = {'max_depth' : range(10, 50, 20), 'random_state':[12345]}
```

```python
model = DecisionTreeRegressor
```

```python
result = my_GridSearchCV(model, parametrs,train_x,train_y)
```

```python
grid_report = grid_report.append(result)
```
